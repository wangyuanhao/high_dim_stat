```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Parameters
n = 2000  # Number of vertices
p = 0.005  # Probability of edge creation

# Generate ErdsRnyi graph
G = nx.erdos_renyi_graph(n, p)

# Plot degree distribution
degrees = [d for _, d in G.degree()]
plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution of ErdsRnyi Graph (n=1000, p=0.01)")
plt.grid()
plt.show()

# Visualize a subset of the graph due to size constraints
subgraph = nx.subgraph(G, np.random.choice(G.nodes, size=200, replace=False))
plt.figure(figsize=(8, 8))
nx.draw(subgraph, node_size=10, edge_color='gray', alpha=0.7)
plt.title("Visualization of a Subgraph of ErdsRnyi Model")
plt.show()
```
