Eigenface

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
```

```python
# Step 1: Load Olivetti Faces dataset
faces_data = fetch_olivetti_faces(shuffle=True)
images = faces_data.images  # Shape (400, 64, 64)
targets = faces_data.target  # Labels (0-39, each person has 10 images)

print(images.shape)
```

```python
# Step 2: Compute Eigenfaces for Each Individual
n_components = 5  # Number of eigenfaces to extract per person


person_id = 2  # Select the first individual


# Select images of this individual
person_images = images[targets == person_id].reshape(10, -1)  # Flatten (10, 4096)

# Apply PCA
pca = PCA(n_components=n_components)
pca.fit(person_images)
```

```python
plt.figure(figsize=(10, 8))
show_images = np.concatenate((pca.components_, person_images), axis=0)  
# Step 3: Visualize Eigenfaces
for j in range(show_images.shape[0]):
    plt.subplot(3, n_components,  j + 1)
    plt.imshow(show_images[j].reshape(64, 64), cmap='gray')
    plt.axis('off')
```
